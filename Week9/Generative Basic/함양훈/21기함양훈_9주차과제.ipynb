```python
# You can freely choose one of architectures you want that looks promising
class Encoder(nn.Module):
    def __init__(self, input_dim, hidden_dim, hidden_dim2, latent_dim):
        super(Encoder, self).__init__()
        self.input1 = nn.Linear(input_dim, hidden_dim)
        self.input2 = nn.Linear(hidden_dim, hidden_dim2)
        self.mean = nn.Linear(hidden_dim2, latent_dim)
        self.var = nn.Linear (hidden_dim2, latent_dim)

        self.LeakyReLU = nn.LeakyReLU(0.2)
        self.training = True

    def forward(self, x):
        h_ = self.LeakyReLU(self.input1(x))
        h_ = self.LeakyReLU(self.input2(h_))
        mean = self.mean(h_)
        log_var = self.var(h_)
        return mean, log_var
```

```python
class Encoder2(nn.Module):
    def __init__(self, input_dim, hidden_dim, hidden_dim2, hidden_dim3, latent_dim):
        super(Encoder2, self).__init__()
        # 추가적인 hidden_layer로 새로운 아키텍처를 정의
        self.input1 = nn.Linear(input_dim, hidden_dim)
        self.input2 = nn.Linear(hidden_dim, hidden_dim2)
        self.input3 = nn.Linear(hidden_dim2, hidden_dim3)
        self.mean = nn.Linear(hidden_dim2, latent_dim)
        self.var = nn.Linear(hidden_dim2, latent_dim)

        # Encoder2에는 ReLU를 사용하여 Encoder1의 LeakyReLU와 비교
        self.ReLU = nn.ReLU()

        self.training = True

    def forward(self, x):
        h_ = self.ReLU(self.input1(x))
        h_ = self.ReLU(self.input2(h_))
        mean = self.mean(h_)
        log_var = self.var(h_)
        return mean, log_var
```

```python
class Decoder(nn.Module):
    def __init__(self, latent_dim, hidden_dim, hidden_dim2, output_dim):
        super(Decoder, self).__init__()
        self.hidden1 = nn.Linear(latent_dim, hidden_dim2)
        self.hidden2 = nn.Linear(hidden_dim2, hidden_dim)
        self.output = nn.Linear(hidden_dim, output_dim)
        self.LeakyReLU = nn.LeakyReLU(0.2)

    def forward(self, x):
        h = self.LeakyReLU(self.hidden1(x))
        h = self.LeakyReLU(self.hidden2(h))
        x_hat = torch.sigmoid(self.output(h))
        return x_hat
```

```python
class Decoder2(nn.Module):
    def __init__(self, latent_dim, hidden_dim3, hidden_dim2, hidden_dim, output_dim):
        super(Decoder2, self).__init__()
        # latent space에서 시작하여 인코더 구조를 반대로 확장
        self.latent_to_hidden = nn.Linear(latent_dim, hidden_dim3)
        self.hidden3_to_hidden2 = nn.Linear(hidden_dim3, hidden_dim2)
        self.hidden2_to_hidden = nn.Linear(hidden_dim2, hidden_dim)
        self.hidden_to_output = nn.Linear(hidden_dim, output_dim)

        # 다양한 activation function 사용
        self.ELU = nn.ELU()

        # 출력에 필요한 경우 최종 레이어에 activation function를 추가하는 것을 고려해야함
        # 예를 들어 출력이 0과 1 사이로 제한된 sigmoid
        self.output_activation = nn.Sigmoid()

    def forward(self, x):
        h_ = self.ELU(self.latent_to_hidden(x))
        h_ = self.ELU(self.hidden3_to_hidden2(h_))
        h_ = self.ELU(self.hidden2_to_hidden(h_))
        output = self.hidden_to_output(h_)
        # 모델의 최종 출력에 필요한 경우 출력 활성화 기능을 적용
        # 예를 들어, MNIST 숫자와 같은 이진 출력의 경우 시그모이드를 사용하여 출력 범위가 [0,1]인지 확인할 수 있음
        output = self.output_activation(output)
        return output
```

```python
class Model(nn.Module):
    def __init__(self, Encoder, Decoder):
        super(Model, self).__init__()
        self.Encoder = Encoder
        self.Decoder = Decoder

    # Think about why you need reparameterization trick (Hint; backpropagation)
    def reparameterization(self, mean, var):
        epsilon = torch.randn_like(var).to(device)
        z = mean + var*epsilon
        return z

    def forward(self, x):
        mean, log_var = self.Encoder(x)
        z = self.reparameterization(mean, torch.exp(0.5 * log_var))
        x_hat = self.Decoder(z)
        return x_hat, mean, log_var
```

```python
# Any modifications are allowed depending on your own Encoder2/Decoder2
# -----------------------------------
x_dim  = 784
hidden_dim = 400
hidden_dim2 = 200
hidden_dim3 = 100
latent_dim = 20
epochs = 30
batch_size = 100
# -----------------------------------

#디바이스를 gpu로 설정
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

encoder = Encoder2(input_dim=x_dim, hidden_dim=hidden_dim, hidden_dim2=hidden_dim2, hidden_dim3=hidden_dim3, latent_dim=latent_dim)
decoder = Decoder2(latent_dim=latent_dim, hidden_dim3=hidden_dim3, hidden_dim2=hidden_dim2, hidden_dim=hidden_dim, output_dim=x_dim)


model = Model(Encoder=encoder, Decoder=decoder).to(device)
```

### Q2. (Optional Assignment)
    Refering to Beta-VAE(https://openreview.net/forum?id=Sy2fzU9gl), give any kind of manipulations on pre-made loss function, and visualize your samples.
    It would be nice to see whether the fidelity of sample data is improved or not! Also guess why.
***Warning)  you don't have to take all the contents of this paper! Look carefully at the Equation (4)***

```python
# (Optional Assignment)
def loss_function(x, x_hat, mean, log_var):
    reproduction_loss = nn.functional.binary_cross_entropy(x_hat, x, reduction='sum')
    # Think about how to calculate KLDivergence in closed form (Hint; assumptions for Gaussian), please refer to the following links
    # https://stats.stackexchange.com/questions/7440/kl-divergence-between-two-univariate-gaussians
    KLD = - 0.5 * torch.sum(1+ log_var - mean.pow(2) - log_var.exp())
    return reproduction_loss, KLD

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
```

```python
def train(epoch, model, train_loader, optimizer):
    model.train()
    train_loss = 0
    for batch_idx, (x, _) in enumerate(train_loader):
        x = x.view(batch_size, x_dim)
        x = x.to(device)

        optimizer.zero_grad()
        x_hat, mean, log_var = model(x)
        BCE, KLD = loss_function(x, x_hat, mean, log_var)
        loss = BCE + KLD

        train_loss += loss.item()
        loss.backward()
        optimizer.step()

        if batch_idx % 100 == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\t Loss: {:.6f}'.format(
                epoch, batch_idx * len(x), len(train_loader.dataset),
                100. * batch_idx / len(train_loader),
                loss.item() / len(x)))

    print("======> Epoch: {} Average loss: {:.4f}".format(
        epoch, train_loss / len(train_loader.dataset)))
```

```python
def test(epoch, model, test_loader):
    model.eval()
    test_loss = 0
    with torch.no_grad():
        for batch_idx, (x, _) in enumerate(test_loader):
            x = x.view(batch_size, x_dim)
            x = x.to(device)
            x_hat, mean, log_var = model(x)
            BCE, KLD = loss_function(x, x_hat, mean, log_var)
            loss = BCE + KLD

            test_loss += loss.item()

            if batch_idx == 0:
                n = min(x.size(0), 8)
                comparison = torch.cat([x[:n], x_hat.view(batch_size, x_dim)[:n]])
                grid = torchvision.utils.make_grid(comparison.cpu())
```

```python
# Trackning the process
# If you have difficulty on installing 'tqdm', modifications are allowed. Not necessary
from tqdm.auto import tqdm

# Training an VAE model, and record losses
for epoch in tqdm(range(0, epochs)):
    train(epoch, model, train_loader, optimizer)
    test(epoch, model, test_loader)
    print("\n")
```
