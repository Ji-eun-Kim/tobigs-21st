```python
#수치형 변수 추출
num_col = data.select_dtypes(include=['float64', 'int64'])
num_col
```

```python
#수치형 변수들의 box plot을 7x4로 출력 - 이산적인 값과 이상치 점을 확인 가능했음.(순서형, 명목형 제외하지 않음)
fig, axes = plt.subplots(nrows=7, ncols=4, figsize=(12, 12))

for i, ax in enumerate(axes.flat):
    if i < len(num_col.columns):
        column = num_col.columns[i]
        ax.boxplot(data[column])
        ax.set_title(f'{column}')
        ax.set_ylabel(column)

plt.tight_layout()
plt.show()
```

```python
#분포 플롯 출력 - 제외할 변수(모든 값이 동일)들과 순서형 자료 의심 가능 및 경향성 파악
fig, axes = plt.subplots(nrows=7, ncols=4, figsize=(15, 20))

for i, ax in enumerate(axes.flat):
    if i < len(num_col.columns):
        column = num_col.columns[i]

        sns.histplot(data[column], ax=ax, kde=True)
        ax.set_title(f'{column}')
        ax.set_xlabel('Values')
        ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
num_col_excluded = data.select_dtypes(include=['float64', 'int64']).drop(['EmployeeCount', 'StandardHours'], axis=1)
```

```python
correlation_matrix = num_col_excluded.corr()

#상관관계 히트맵
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('correlation')
plt.show()
```

```python
obj_col = data.select_dtypes(include='object')
obj_col 
```

```python
#범주형 변수들의 bar plot
plt.figure(figsize=(12, 8))

fig, axes = plt.subplots(3, 3, figsize=(12, 12))

for i, column in enumerate(obj_col.columns):
    row = i // 3
    col = i % 3
    ax = axes[row, col]
    sns.countplot(x=column, data=obj_col, palette="viridis", ax=ax)
    ax.set_title(f'{column}')

plt.tight_layout()
plt.show()
```

```python
#순서형 변수에 대한 파이 차트
plt.figure(figsize=(15, 8))

#EnvironmentSatisfaction
plt.subplot(2, 3, 1)
data['EnvironmentSatisfaction'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Environment Satisfaction')

#JobInvolvement
plt.subplot(2, 3, 2)
data['JobInvolvement'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Job Involvement')

#PerformanceRating
plt.subplot(2, 3, 3)
data['PerformanceRating'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Performance Rating')

#RelationshipSatisfaction
plt.subplot(2, 3, 4)
data['RelationshipSatisfaction'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Relationship Satisfaction')

#WorkLifeBalance
plt.subplot(2, 3, 5)
data['WorkLifeBalance'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Work-Life Balance')

plt.tight_layout()
plt.show()
```

```python
#파생변수 생성-스트레스 지수
stress_columns = ['EnvironmentSatisfaction', 'JobSatisfaction', 'WorkLifeBalance']
data['Stress_Index'] = (data[stress_columns].sum(axis=1) / data['DailyRate']).fillna(0)
```
