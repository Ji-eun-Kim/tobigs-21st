# Outliers of Numerical Data

```python
# Distinguish Numerical variables and Categorical variables
num_data = data._get_numeric_data() # Numerical data
# num_data = data.select_dtypes(exclude=object)
```

## 1. Outliers by Z-score

```python
def outlier_by_Zscore(data):
  # z_score 구하기
  mean = np.mean(data)
  std = np.std(data)
  z_scores = (data - mean) / std

  # outlier로 처리할 임계 z_score 값을 3으로 설정
  # outlier 확인
  z_score_outliers = z_scores[(z_scores > 3) | (z_scores < -3)]

  return z_score_outliers
```

```python
outliers_Zscore = outlier_by_Zscore(num_data)

outliers_Zscore.info()
```

## 2. Outliers by IQR


```python
# Outlier 시각적으로 확인
plt.figure(figsize=(20, 30))
for i in range(1, len(num_data.columns.to_list())+1):
  plt.subplot(4, 7, i)
  num_data.boxplot(column = num_data.columns.to_list()[i-1])
```

```python
# outliers by IQR 확인
def outliers_by_IQR(data):
  Q1 = data.quantile(q=0.25)
  Q3 = data.quantile(q=0.75)
  IQR = Q3 - Q1
  IQR_outliers = num_data[(num_data < Q1-1.5*IQR) | (num_data > Q3+1.5*IQR)]

  return IQR_outliers

outliers_IQR = outliers_by_IQR(num_data)
outliers_IQR.info()
```

Z_score보다 IQR 기준으로 했을 때, 더 많은 변수에 대해 더 많은 이상치를 탐지할 수 있었다.

## Remove the Instances with the Outliers


```python
# Outliers by Zscore가 존재하는 인스턴스를 추출
outliers = outliers_Zscore.dropna(how='all', axis=0) # 모든 값이 NaN인 행(인스턴스)을 제거함으로써 이상치가 존재하는 인스턴스만을 추출

# Remove the rows with Outliers by Z_score
cleaned_data = data.drop(index = outliers.index)
```

## Distribution of the values based on Attrition: Barplot


```python
independent_data = cleaned_data.drop(columns='Attrition') # 독립변수

# Calculate the number of rows and columns for subplots
num_rows = (len(independent_data.columns) // 2) + (len(independent_data) % 2 > 0)  # Round up division
num_cols = 2

# Create subplots
fig, axs = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(20, 100), sharex=False,
                        gridspec_kw={'hspace': 0.5, 'wspace': 0.3})

# Flatten the axs array to simplify indexing
axs = axs.flatten()

for i in range(len(axs)):
    if i < len(independent_data.columns):
      if independent_data[independent_data.columns[i]].dtype == object:
        sns.countplot(data=cleaned_data, x=independent_data.columns[i], hue='Attrition', ax=axs[i])
        axs[i].tick_params(axis='x', rotation=45)
      else:
          sns.barplot(data=cleaned_data, x='Attrition', y=independent_data.columns[i], ax=axs[i])
          axs[i].tick_params(axis='x', rotation=45)
```
