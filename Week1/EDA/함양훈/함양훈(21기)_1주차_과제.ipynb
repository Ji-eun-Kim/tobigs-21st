##**Check Outlier**

```python
import seaborn as sns

#수치형 변수 선택
numeric_columns = data.select_dtypes(include=['int64', 'float64']).columns

#multi boxplot으로 시각화
plt.figure(figsize=(18,9))
sns.boxplot(data=data[numeric_columns])
plt.xticks(rotation=90)
plt.title('Boxplot for Numeric Variables')
plt.show()
```

```python
#IQR 방법으로 이상치 탐지
Q1 = data[numeric_columns].quantile(0.25)
Q3 = data[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

#이상치로 간주되는 데이터 포인트를 식별 (1.5 * IQR 규칙 사용)
outliers = ((data[numeric_columns] < (Q1 - 1.5 * IQR)) | (data[numeric_columns] > (Q3 + 1.5 * IQR))).sum()
outliers[outliers > 0]  #이상치가 하나 이상 있는 변수들만 출력
```

#**Scaling**

```python
from sklearn.preprocessing import RobustScaler

#RobustScaler
robust_scaler = RobustScaler()

#수치형 변수만 선택
numeric_data = data.select_dtypes(exclude='object')
robust_scaled_data = robust_scaler.fit_transform(numeric_data)

# Robust 스케일링된 데이터를 DataFrame으로 변환
robust_scaled_data_df = pd.DataFrame(robust_scaled_data, columns=numeric_data.columns)
robust_scaled_data_df.head()
```

#**Heatmap correlation**

```python
1#상관계수 행렬 계산
corr_matrix = robust_scaled_data_df.corr()

#히트맵
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#상관계수 계산
numerical_data = robust_scaled_data_df.select_dtypes(include=np.number)
corr_matrix = numerical_data.corr()

#상관계수가 0.75 이상
highly_significant_corr_pairs = {}
threshold = 0.75
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        corr = corr_matrix.iloc[i, j]
        if abs(corr) >= threshold:
            var1, var2 = corr_matrix.columns[i], corr_matrix.columns[j]
            highly_significant_corr_pairs[(var1, var2)] = corr


highly_significant_corr_pairs
```

#**Feature Engineering**

```python
#연령 그룹
bins = [18, 29, 39, 49, 59, 100]  #범주 설정
labels = ['20s', '30s', '40s', '50s', '60s+']  #각 그룹에 레이블 지정
data['AgeGroup'] = pd.cut(data['Age'], bins=bins, labels=labels)

data['AgeGroup'].head()
```

```python
#시간당 급여 대비 월 소득 비율
data['HourlyRateRatio'] = data['HourlyRate'] / (data['MonthlyIncome'] / 160)  #월 소득을 시간당 급여로 나눔


data['HourlyRateRatio'].head()
```

```python
#근속 연수 대비 프로모션 빈도
data['PromotionFrequency'] = data['YearsSinceLastPromotion'] / (data['YearsAtCompany'] + 1)  #0으로 나누는 경우를 방지하기 위해 +1 추가

data['PromotionFrequency'].head()
```
