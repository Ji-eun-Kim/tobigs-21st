### 1. Normal Equation

```python
import numpy as np
from numpy.linalg import inv
```

```python
def estimate_beta(X, y):
    designX = np.hstack([np.ones((X.shape[0], 1)), X]) 
    # 첫 번째 열을 1로 채운 열 벡터와 원래의 X를 수평 방향으로 병합
    beta_hat = np.linalg.inv(designX.T.dot(designX)).dot(designX.T).dot(y) 
    # 정규방정식은 (X^T * X)^-1 * X^T * y 형태
    # inv 함수를 사용하여 역행렬을 계산하고, dot 함수를 사용하여 행렬의 곱셈을 수행

    return beta_hat
```

```python
beta_hat = estimate_beta(X, y)
beta_hat
```

### 2. MSE

```python
def MSE(X, y, beta_hat):
    designX = np.hstack([np.ones((X.shape[0], 1)), X]) 
    y_pred = designX.dot(beta_hat)
    mse = np.mean((y - y_pred)**2)

    return mse
```

```python
MSE(X, y, estimate_beta(X, y))
```
