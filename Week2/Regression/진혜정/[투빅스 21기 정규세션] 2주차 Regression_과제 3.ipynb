## Model

```python
from sklearn.linear_model import LogisticRegression

y = data['Class']
X = data.drop('Class', axis = 1)

model = LogisticRegression()
model.fit(X, y)

coef = model.coef_[0][0]
intercept = model.intercept_[0]

print(f"Logistic Regression Equation: log(odds) = {coef:.2f}x + {intercept:.2f}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

predictions = model.predict(X)
print("Accuracy:", round(accuracy_score(y, predictions),4))
print("Confusion Matrix:")
print(confusion_matrix(y, predictions))
print("Classification Report:")
print(classification_report(y, predictions))
```

Accuracy : 성능지표가 매우 높지만, 데이터가 불균형하여 나온 값이므로 신뢰하기 어렵다.

Confustion matrix : 특정 값들이 매우 적은수로 있는 것을 확인할 수 있다.

Classification Report
- Precision : 1이라고 예측한 것 중에 실제 1을 맞출 확률 95%
- Recall : 실제 1 값을 올바르게 예측한 확률 80%
- F1-score : Precision & Recall의 조화평균이 87%
- Support : 데이터의 불균형 확인


Recall과 Support을 개선하기 위해 적은 0의 데이터 수를 고려하는 parameter 추가

```python
model2 = LogisticRegression(max_iter=1000, class_weight='balanced')
```

```python
model2.fit(X, y)
predictions = model2.predict(X)

print("Accuracy:", round(accuracy_score(y, predictions),4))
print("Confusion Matrix:")
print(confusion_matrix(y, predictions))
print("Classification Report:")
print(classification_report(y, predictions))
```
