### 1. Normal Equation

```python
import numpy as np
from numpy.linalg import inv
```

```python
def estimate_beta(X, y):

    bias = np.ones((len(y),1))
    designX = np.concatenate([bias, X], axis = 1)
    beta_hat = inv(np.dot(designX.T,designX)) @ designX.T @ y

    return beta_hat
```

```python
beta_hat = estimate_beta(X, y)
beta_hat
```

### 2. MSE

```python
def MSE(X, y, beta_hat):
    designX = np.concatenate([np.ones((len(y), 1)), X], axis=1)
    y_pred = designX @ beta_hat
    mse = np.sum(np.power(y - y_pred,2))
    mse = mse / (len(y) - (X.shape[1] + 1))

    return mse
```

```python
MSE(X, y, estimate_beta(X, y))
```
